```python
#PRE PROCESSING EXERCISE FOR WEEK 5 - SARCAUGA
#IMPORTS MACHINE LEARNING
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector
from sklearn.impute import SimpleImputer

#Read the path
path = '/content/Raw_Medical_Data_for_day1.csv'
df = pd.read_csv(path)
df.head()
```

```python
# Train test split the data to prepare for machine learning

X = df.drop(columns = ['Unnamed: 0', 'Additional_charges'])
y = df['Additional_charges']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

#make categorical and numeric selectors
cat_selector = make_column_selector(dtype_include='object')

#select categorical columns
cat_data = X_train[cat_selector(X_train)]
cat_data
```

```python
# One Hot Encode the nominal features
from sklearn.compose import make_column_selector
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
region_selector = make_column_selector(dtype_include='object')
region_data = X_train[region_selector(X_train)]
region_data
ohe_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
ohe_encoder.fit(region_data)
region_ohe = ohe_encoder.transform(region_data)
region_ohe
pd.DataFrame(region_ohe, columns=ohe_encoder.get_feature_names(region_data.columns))
```
